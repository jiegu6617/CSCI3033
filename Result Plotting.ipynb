# Model Comparison

```python
xception = pd.read_csv('xception_adam_0001.csv')
xception_no = pd.read_csv('xception_noaugmentation.csv')
vgg = pd.read_csv('train_final_vgg.csv')
res50 = pd.read_csv('res_50_with_augment.csv')
vgg_bn = pd.read_csv('vgg_bn.csv')
```

```python
plt.figure(figsize=(8,6))
plt.plot(xception['val_accuracy'][:50],label = 'xception with data augmentation')
plt.plot(xception_no['val_accuracy'][:50],label = 'xception without data augmentation')
plt.plot(vgg['val_accuracy'][:50],label = 'vgg16')
plt.plot(res50['val_accuracy'][:50],label = 'resnet50')
plt.plot(vgg_bn['val_accuracy'][:50],label = 'vgg_bn (no transfer learning)')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.title('Model Comparison')

plt.legend()
plt.show()
```

```python
avg_time = [round(xception['time'].mean(),2),round(xception_no['time'].mean(),2),round(vgg['time'].mean(),2),
            round(vgg['time'].mean(),2),round(vgg_bn['time'].mean(),2)]
models = ['Xception aug','Xception','vgg16','resnet50','vgg_bn']

plt.figure(figsize=(8,6))
plt.bar(models,avg_time)
plt.ylabel('Time (s)')
plt.title('Model Time Comparison')
plt.show()
```

# Gpu Comparison

```python
price = [0.45,1.46,2.48]
gpus = ['K80','P100','V100']

plt.figure(figsize=(8,6))
plt.bar(gpus,price)
plt.ylabel('$/hour')
plt.title('GPU Price Comparision')
plt.show()
```

```python
xception_v100 = pd.read_csv('xception_v100.csv')
xception_p100 = pd.read_csv('xception_p100.csv')
xception_k80 = pd.read_csv('xception_k80.csv')
```

```python
times = [round(xception_k80['time'].mean(),2),round(xception_p100['time'].mean(),2),round(xception_v100['time'].mean(),2)]

plt.figure(figsize=(8,6))
plt.bar(gpus,times)
plt.ylabel('s')
plt.title('GPU Runtime Comparision')
plt.show()
```
